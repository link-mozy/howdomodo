```python
from pyspark import SparkContext
from pyspark.sql import SparkSession

sc = SparkContext()
```

```python
spark = SparkSession.builder \
    .master("local") \
    .appName("Word Count") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
```

```python
df = spark.read.option("delimeter", ",").option('header',True).csv('2019-09.csv')
df.collect()
```

```python
df.count()
```

```python
df
```

```python
rdd = sc.parallelize(df.collect())
```

```python
type(rdd.collect())
```

```python
rdd.collect()[:3]
```

```python
from pyspark.sql import Row
```

```python
rdd = rdd.map(lambda x: Row(성별=int(x[0][0]),\
                      연령=int(x[1][0]),\
                      승인시간대1 = int(x[2]),\
                      가맹점업종코드 =int(x[3]) ))
```

```python
rdd.collect()[:3]
```

```python
from pyspark.mllib.util import MLUtils
from pyspark.mllib.regression import LabeledPoint
```

```python
print(rdd.take(3))
c = rdd
```

```python
d = c.map(lambda line: LabeledPoint(line[3],[line[:3]])) # arbitrary mapping, it's just an example
```

```python
print(d.take(3))
```

```python
MLUtils.saveAsLibSVMFile(d, "test")
```

```python
from pyspark.ml.clustering import GaussianMixture
```

```python
# loads data
dataset = spark.read.format("libsvm").load("2019-09.txt")
```

```python
gmm = GaussianMixture().setK(2).setSeed(538009335)
model = gmm.fit(dataset)

print("Gaussians shown as a DataFrame: ")
model.gaussiansDF.show(truncate=False)
```
