```python
headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
url = "https://movie.naver.com/movie/running/current.nhn"
```

```python
def download(method, url, headers=None, params=None, data=None, timeout=1, maxretries=3):
    """requests를 이용하여 download하는 함수"""
    try:
        resp = requests.request(method, url, params=params, data=data, headers=headers)
        resp.raise_for_status()
    except requests.exceptions.HTTPError as e:
        if 500 <= e.status_code() < 600 and maxretries > 0:
            print(maxretries)
            time.sleep(timeout)
            download(method, url, headers, params, data, timeout, maxretries-1)
        print(e.response.status_code, e.response.reason)
    return resp
```

```python
def parser(method, url, params, headers):
    resp = requests.request(method, url, params=params, data=None, headers=headers)
    dom = BeautifulSoup(resp.text, 'html5lib')
    z1hOCe = dom.select_one('.Z1hOCe')
    lrzXr = dom.select_one(".LrzXr")
    return lrzXr.text
```

```python
def getTheaterAddress(theater_list):
    columns = ['theater_name', 'theater_address']
    result = pd.DataFrame(columns=columns);
    url = "https://www.google.co.kr/search?"
    headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
    
    for theater_name in theater_list:
        params = {"q" : theater_name}
        theater_address = parser('get', url, params, headers)
        print("name: " + theater_name + ", address: " + theater_address)
        result.append({'theater_name':theater_name, 'theater_address':theater_address}, ignore_index=True)
    
    return result
```
