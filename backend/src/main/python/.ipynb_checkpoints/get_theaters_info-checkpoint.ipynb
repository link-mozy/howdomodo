```python
# 주소 파싱

def parser(method, url, params, headers):
    try:
        resp = requests.request(method, url, params=params, data=None, headers=headers)
        dom = BeautifulSoup(resp.text, 'html5lib')
        z1hOCe = dom.select_one('.Z1hOCe')
        lrzXr = dom.select_one(".LrzXr")
        result = lrzXr.text
    except Exception as e:
        print(e)
        result = ''
    return result
```

```python
# 주소 구하기

def getTheaterAddress(theater_list):
    columns = ['theater_name', 'theater_address']
    result = pd.DataFrame(columns=columns);
    url = "https://www.google.co.kr/search?"
    headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
    start = 0
    end = len(theater_list)
    
    for theater_name in theater_list:
        params = {"q" : theater_name}
        theater_address = parser('get', url, params, headers)
        result = result.append({'theater_name':theater_name, 'theater_address':theater_address}, ignore_index=True)
        start += 1
        print("{}/{}. name: {}, address: {}".format(start, end, theater_name, theater_address))
    
    return result
```

```python
theater_address = getTheaterAddress(theaters)
```

```python
# 위도,경도 구하기
def getLatLon(address, reg):
    try:
        url = "https://www.google.co.kr/maps/place/" + address
        print(url)
        headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
        resp = requests.request('get', url)
        dom = BeautifulSoup(resp.text, 'html5lib')
        meta_list = dom.find('head').select('meta')

        lat_lon = []
        for content in meta_str_list:
            m = reg.match(content)
            if m:
                lat_lon.append(m.group(2))

        lat_lon = lat_lon[0].split('%2C')
        
    except Exception as e:
        print(e)
        lat_lon = ['', '']
        
    return lat_lon
```

```python
def naveropenapiGeocode(search_address):
    headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36", 
              "X-NCP-APIGW-API-KEY-ID": "bril3ppp5c",
              "X-NCP-APIGW-API-KEY": "hSwItsYFcmb3BTKGSrP2QDvMALuMMSMhO9JLzkA7"}
    url = "https://naveropenapi.apigw.ntruss.com/map-geocode/v2/geocode?"
    params = {"query" : search_address}
    resp = requests.request('get', url, params=params, data=None, headers=headers)
    addresses = resp.json()['addresses']
    return addresses
```

```python
def getLatLon(addresses):
    result = {}
    if len(addresses) == 1 :
        # 바로 저장
        address_ = addresses[0]
        jibunAddress = address_['jibunAddress']
        # 위도(latitude : y) 경도(longitude : x)
        lat = address_["y"]
        lon = address_["x"]

        result['jibunAddress'] = jibunAddress
        result['lat'] = lat
        result['lon'] = lon
    
    elif len(addresses) > 1:
        # 맞는거 찾아서 저장
        point_list = []

        for address_ in addresses:
            addressElements = address_['addressElements']
            point = 0

            for addressElement in addressElements:
                longName = addressElement['longName']
                shortName = addressElement['shortName']
                if any(elem in address_split for elem in [longName, shortName]):
                    point += 1

            point_list.append(point)

        point_max_index = point_list.index(max(point_list))
        address_ = addresses[point_max_index]
        jibunAddress = address_['jibunAddress']
        # 위도(latitude : y) 경도(longitude : x)
        lat = address_["y"]
        lon = address_["x"]

        result['jibunAddress'] = jibunAddress
        result['lat'] = lat
        result['lon'] = lon
        
    return result
```

```python
def getTheaterLotLonByAddress(df_theater_address):
    columns = ["theater_name", "theater_address", "jibunAddress", "theater_lat", "theater_lon"]
    result = pd.DataFrame(columns=columns)
    total = len(df_theater_address)
    
    for index in range(total):
        theater_address = df_theater_address.iloc[index].theater_address
        theater_name = df_theater_address.iloc[index].theater_name
        new_row = {}
        new_row["theater_address"] = theater_address
        new_row["theater_name"] = theater_name
        new_row["jibunAddress"] = ''
        new_row["theater_lat"] = ''
        new_row["theater_lon"] = ''

        addresses = None

        for number in range(2,5):
            address_split = theater_address.split(',')[0].split()
            search_address = ' '.join(address_split[-number:])
            addresses = naveropenapiGeocode(search_address)
            if len(addresses) > 0:
                break

        if addresses is not None:
            latlon_dic = getLatLon(addresses)
            if len(latlon_dic.items()) > 0:
                new_row["jibunAddress"] = latlon_dic["jibunAddress"]
                new_row["theater_lat"] = latlon_dic["lat"]
                new_row["theater_lon"] = latlon_dic["lon"]

        result = result.append(new_row, ignore_index=True)
        
    return result
```

```python
theater_address_lat_lon = getTheaterLotLonByAddress(theater_address)
theater_address_lat_lon
```
