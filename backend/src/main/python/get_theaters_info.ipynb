```python
# 라이브러리
import re
import requests
from bs4 import BeautifulSoup
import time
import json
from selenium import webdriver
from html import unescape

import pandas as pd
import numpy as np
```

```python
# 영화관 이름

theaters = pd.read_csv('test.csv', header=None)
theaters = theaters[0].tolist()
theaters
```

```python
# 주소 파싱

def parser(method, url, params, headers):
    try:
        resp = requests.request(method, url, params=params, data=None, headers=headers)
        dom = BeautifulSoup(resp.text, 'html5lib')
        z1hOCe = dom.select_one('.Z1hOCe')
        lrzXr = dom.select_one(".LrzXr")
        result = lrzXr.text
    except Exception as e:
        print(e)
        result = ''
    return result
```

```python
# 주소 구하기

def getTheaterAddress(theater_list):
    columns = ['theater_name', 'theater_address']
    result = pd.DataFrame(columns=columns);
    url = "https://www.google.co.kr/search?"
    headers = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
    start = 0
    end = len(theater_list)
    
    for theater_name in theater_list:
        params = {"q" : theater_name}
        theater_address = parser('get', url, params, headers)
        result = result.append({'theater_name':theater_name, 'theater_address':theater_address}, ignore_index=True)
        start += 1
        print("{}/{}. name: {}, address: {}".format(start, end, theater_name, theater_address))
    
    return result
```

```python
theater_address = getTheaterAddress(theaters)
```
